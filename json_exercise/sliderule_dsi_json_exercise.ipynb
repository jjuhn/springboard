## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# 1. Find the 10 countries with most projects
pd.set_option('display.max_colwidth', -1)

json_df = pd.read_json('data/world_bank_projects.json')
json_df.groupby(['countryname']).size()

json_df.countryname.value_counts().sort_values(ascending=False).head(10)
```

```python
# Looking at the result set, I found "Africa" being in a countryname. 
json_df[json_df.countryname=='Africa'].head(10)
```

```python
# further researching into the African country, in the "borrowers" field, there were some information 
# about who borrowed but not necessarily a country.   
# I could use "regionname" to remove regions in "countryname" 

regions = json_df.regionname.unique()
json_df[~json_df.countryname.isin(regions)].countryname.value_counts().sort_values(ascending=False).head(10)
```

```python
#2. Find the top 10 major project themes (using column 'mjtheme_namecode')
#3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

import json

# reading from raw json file and normalize by mjtheme_namecode. 
with open('data/world_bank_projects.json') as json_data:
    d = json.load(json_data)
    df = json_normalize(d, "mjtheme_namecode", ['countryname', 'project_name', 'id'])


s1 = df.name.value_counts().head(10)
raw_df = pd.DataFrame({'name':s1.index, 'counts':s1.values})
# In raw_df, Found missing 122 "name" counts  
raw_df
```

```python
# Created a lookup table to merge. 
lookup = df.groupby(['code', 'name']).size().to_frame().reset_index()
lookup = lookup.rename(columns= {0: 'count'})
lookup = lookup[~(lookup.name=="")]

# merge lookup table and clean up.  Here I am simply assuming that 
df2 = df.merge(lookup, on="code")
df2 = df2.drop(["name_x", "count"], axis=1).rename(columns= {"name_y": "name"})
s2 = df2.name.value_counts().head(10)

cleaned_df = pd.DataFrame({'name': s2.index, 'counts': s2.values})
cleaned_df
```
